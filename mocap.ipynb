```

def get_joint_pos_dict(c_joints, c_motion):
    c_joints['root'].set_motion(c_motion)
    out_dict = {}
    for k1, v1 in c_joints['root'].to_dict().items():
        for k2, v2 in zip('xyz', v1.coordinate[:, 0]):
            out_dict['{}_{}'.format(k1, k2)] = v2
    return out_dict
motion_df = pd.DataFrame([get_joint_pos_dict(joints, c_motion) for c_motion in motions])
motion_df.to_csv('motion.csv', index=False)
motion_df.sample(100)
```

```
fig, m_axs = plt.subplots(3, 1, figsize=(20, 10))
for c_x, c_ax in zip('xyz', m_axs):
    for joint_name in joints.keys():
        if ('hipjoint' in joint_name)or('hand' in joint_name)or ('thorax' in joint_name):
            c_ax.plot(motion_df['{}_{}'.format(joint_name, c_x)], label=joint_name)
            c_ax.set_title(c_x);
            m_axs[0].legend()
    plt.xlabel("timestamps")
```

```
lfoot_x=motion_df['lhipjoint_x'][4000:4160]
rfoot_x=motion_df['rhipjoint_x'][4000:4160]
thorax_x=motion_df['thorax_x'][4000:4160]
right_hand_x=motion_df['rhand_x'][4000:4160]
left_hand_x=motion_df['lhand_x'][4000:4160]


pi = np.pi
  # Radar parameters
fs =40e8# Sampling frequency in Hz, i.e., we sample 78125
  # times per second
ts = 1 / fs # Sampling time, i.e., one sample is taken each
  # ts seconds
Teff =2048 * ts # Total sampling time for 2048 samples
  # (AKA effective sweep duration) in seconds.
Beff =1.5e9# Range of transmit signal frequency during the time the
  # radar samples, known as the "effective bandwidth"
  # (given in Hz)
S =Beff/Teff # Frequency sweep rate in Hz/
P = np.array([0, pi / 2, pi / 3, pi / 5, pi / 6]) # Randomly chosen phase offsets
t = np.arange(len(lfoot_x)) * ts # Sample times
R = np.array([100, 137, 154, 159, 180])
fd2x = R[0]* lfoot_x 
fd3x = R[1]* rfoot_x 
fd4x = R[2]* thorax_x 
fd5x = R[3]* right_hand_x
fd6x=R[4]*left_hand_x

M = np.array([0.9, 0.9, 0.9, 0.7, 0.7])


# Generate five targets
v0x=M[0]* fd2x
v1x= M[1]*fd3x
v2x= M[2]*fd4x
v3x= M[3]*fd5x
v4x=M[4]*fd6x



v_sumx=v0x+v1x+v2x+v3x+v4x
v_sumx=np.conj(v_sumx)
```

```
lfoot_y=motion_df['lhipjoint_y'][4000:4160]
rfoot_y=motion_df['rhipjoint_y'][4000:4160]
thoray_y=motion_df['thorax_y'][4000:4160]
right_hand_y=motion_df['rhand_y'][4000:4160]
left_hand_y=motion_df['lhand_y'][4000:4160]


pi = np.pi
  # Radar parameters
fs =40e8 # Sampling frequency in Hz, i.e., we sample 78125
  # times per second
ts = 1 / fs # Sampling time, i.e., one sample is taken each
  # ts seconds
Teff =2048 * ts # Total sampling time for 2048 samples
  # (AKA effective sweep duration) in seconds.
Beff =1.5e9# Range of transmit signal frequency during the time the
  # radar samples, known as the "effective bandwidth"
  # (given in Hz)
S =Beff/Teff # Frequency sweep rate in Hz/
P = np.array([0, pi / 2, pi / 3, pi / 5, pi / 6])# Randomly chosen phase offsets
t = np.arange(len(lfoot_y)) * ts # Sample times
M = np.array([0.9, 0.9, 0.9, 0.7, 0.7])
R = np.array([100, 137, 154, 159, 180])             
fd2y = R[0]* lfoot_y 
fd3y = R[1]* rfoot_y 
fd4y = R[2]* thoray_y 
fd5y = R[3]* right_hand_y
fd6y=R[4]*left_hand_y


# Generate five targets
v0y=M[0]* fd2y
v1y= M[1]*fd3y
v2y= M[2]*fd4y
v3y= M[3]*fd5y
v4y=M[4]*fd6y


v_sumy=v0y+v1y+v2y+v3y+v4y
v_sumy=np.conj(v_sumy)
```
